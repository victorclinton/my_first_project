```python
"""
Activation Functions Visualizer
--------------------------------
Defines and plots three common neural network activation functions:
  1. Sigmoid
  2. Tanh
  3. ReLU

x values range from -20 to 10, and y values are the transformed outputs
from each activation function.
"""

import numpy as np
import matplotlib.pyplot as plt

 
# ---------------------------------------------------------
# 1. Activation function formulas
# ---------------------------------------------------------
def sigmoid(x):
    """Sigmoid: f(x) = 1 / (1 + e^-x). Output range: (0, 1)."""
    return 1 / (1 + np.exp(-x))


def tanh(x):
    """Tanh: f(x) = (e^x - e^-x) / (e^x + e^-x). Output range: (-1, 1)."""
    return np.tanh(x)


def relu(x):
    """ReLU: f(x) = max(0, x). Output range: [0, inf)."""
    return np.maximum(0, x)


# ---------------------------------------------------------
# 2. Generate input values
# ---------------------------------------------------------
x = np.linspace(-20, 10, 500)

y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)


# ---------------------------------------------------------
# 3. Plot each function side by side
# ---------------------------------------------------------
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

plot_data = [
    ("Sigmoid", y_sigmoid, "tab:blue"),
    ("Tanh", y_tanh, "tab:green"),
    ("ReLU", y_relu, "tab:red"),
]

for ax, (name, y, color) in zip(axes, plot_data):
    ax.plot(x, y, color=color, linewidth=2)
    ax.axhline(0, color="black", linewidth=0.6)
    ax.axvline(0, color="black", linewidth=0.6)
    ax.set_title(f"{name} Activation Function")
    ax.set_xlabel("x")
    ax.set_ylabel(f"{name}(x)")
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
plt.savefig("activation_functions_separate.png", dpi=150)
```

```python
# 4. Plot all three together for direct comparison
# ---------------------------------------------------------
plt.figure(figsize=(9, 6))
plt.plot(x, y_sigmoid, label="Sigmoid", color="tab:blue", linewidth=2)
plt.plot(x, y_tanh, label="Tanh", color="tab:green", linewidth=2)
plt.plot(x, y_relu, label="ReLU", color="tab:red", linewidth=2)
plt.axhline(0, color="black", linewidth=0.6)
plt.axvline(0, color="black", linewidth=0.6)
plt.title("Comparison of Activation Functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("activation_functions_comparison.png", dpi=150)

plt.show()
```

```python
import pandas as pd

# ---------------------------------------------------------
# 2. Visualise with a grouped bar chart
# ---------------------------------------------------------
data = {
    "Model": ["Linear Regression","Lasso Regression","Ridge Regression"],
    "MAE": [0.49,0.78,0.81],
    "MSE": [0.45,0.79,0.82],
    "RMSE": [0.68,0.81,0.85]
}
df = pd.DataFrame(data)
metrics = ["MAE", "MSE", "RMSE"]
models = df["Model"].tolist()

x = np.arange(len(models))      # positions for each model group
width = 0.25                    # width of each bar

fig, ax = plt.subplots(figsize=(9, 6))

colors = ["tab:blue", "tab:orange", "tab:green"]

for i, metric in enumerate(metrics):
    offset = (i - 1) * width
    bars = ax.bar(x + offset, df[metric], width, label=metric, color=colors[i])
    # annotate each bar with its value
    ax.bar_label(bars, fmt="%.2f", padding=2, fontsize=9)

ax.set_xlabel("Model")
ax.set_ylabel("Error Value")
ax.set_title("Performance Comparison of Regression Models")
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend(title="Metric")
ax.grid(True, axis="y", alpha=0.3)
plt.tight_layout()
plt.savefig("model_performance_comparison.png", dpi=150)
plt.show()
```
